## 式一覧
上から順に計算する

- $W_f=(1-\lambda_w)W_{f-1}+\lambda_w\delta(C_{-1}+C_{a-1}+I_{-1}+G_{-1}-T_{v-1}-T_{af-1}-r_LL_{f-1})$
- $W_b=(1-\lambda_w)W_{b-1}+\lambda_w\delta(-T_{cb-1}+P_{b-1}-TD_{b-1}+r_LL_{-1}+r_{GB}GB_{b-1})$
- $W_g=W_{g-1}(1+\gamma)$ (外生的に決める場合)
- $W_g=(1-\lambda_g)(T_{i-1}+T_{a-1}+T_{c-1}+T_{v-1}+T_{f-1}-r_{GB}(GB_{-1}-\Delta GB_{-1}))(1+\gamma)$ (内生的に決める場合)
- $W=W_f+W_b+W_g$
- $T_{iw}=\epsilon_1 W$
- $T_{ci}=\max\{0, \epsilon_7\Delta p_{e-1}(e_{i-1}-\Delta e_{i-1})+\epsilon_9(E_{i-1}+D_{-1})\}$
- $T_{cb}=\max\{0, \epsilon_7\Delta p_{e-1}(e_{b-1}-\Delta e_{b-1})+\epsilon_9E_{b-1}\}$
- $T_c=T_{ci}+T_{cb}$
- $T_{aw}=\epsilon_5K_{w-1} + \epsilon_8(M_{w-1}-L_{i-1})$
- $T_{ai}=\epsilon_5K_{i-1} + \epsilon_8(M_{i-1}+E_{i-1}+GB_{i-1})$
- $T_{af}=\epsilon_6K_{f-1}$
- $T_a=T_{aw}+T_{ai}+T_{af}$
- $G=G_{g-1}(1+\gamma)$ (外生的に決める場合)
- $G=\lambda_g(T_{i-1}+T_{a-1}+T_{c-1}+T_{v-1}+T_{f-1}-r_{GB}(GB_{-1}-\Delta GB_{-1}))(1+\gamma)$ (内生的に決める場合)
- $C_w=\alpha_4\{\alpha_1(W-T_{iw}-T_{aw}-r_LL_{w-1})+\alpha_2(NW_{w-1}-K_{w-1})\}$
- $C_i=\alpha_5\{\alpha_3(P_{i-1}+TD_{b-1}-T_{ii-1}-T_{ai-1}-T_{ci}+r_{GB}GB_{i-1})+\alpha_6(M_{i-1}+E_{i-1}+D_{-1}+GB_{i-1})\}$
- $C=C_w+C_i$
- $C_{aw}=(1-\alpha_4)\{\alpha_1(W-T_{iw}-T_{aw}-r_LL_{w-1})+\alpha_2(NW_{w-1}-K_{w-1})\}$
- $C_{ai}=(1-\alpha_5)\{\alpha_3(P_{i-1}+TD_{b-1}-T_{ii-1}-T_{ai-1}-T_{ci}+r_{GB}GB_{i-1})+\alpha_6(M_{i-1}+E_{i-1}+D_{-1}+GB_{i-1})\}$
- $C_a=C_{aw}+C_{ai}$
- $I=\max[0, \min\{\beta_5(M_{f-1}-L_{f-1})+(\frac{C+C_a+G}{\beta_1 K_{f-1}}-u^T)K_{f-1}+\beta_2K_{f-1}, \beta_3(M_{f-1}-L_{f-1})\}]$
- $T_v=\epsilon_2(C+C_a+G+I)$
- $T_{ff}=\max\{0,\epsilon_3(C+C_a+G+I-W_f-T_v-T_{af}-r_LL_{f-1})\}$
- $P=C+C_a+G+I-W_f-T_v-T_{ff}-T_{af}-r_LL_{f-1}$
- $TD_f = \max\{0, \zeta_1(P-I)+\zeta_2(M_{f-1}-L_{f-1})\}$
- $P_i=TD_f\frac{E_{i-1}}{E_{-1}}$
- $P_b=TD_f\frac{E_{b-1}}{E_{-1}}$
- $P_f=P-P_i-P_b$
- $TD_b=\max\{0, \theta_1 (-M_{-1}+L_{-1}+E_{b-1}+GB_{b-1}+H_{b-1})+ \theta_2 NL_{b-1}\}$
- $T_{ii}=\epsilon_4(P_i+TD_b+r_{GB}GB_{i-1})$
- $T_i=T_{iw}+T_{ii}$
- $T_{fb}=\max\{0, \epsilon_3(-W_b+P_b+r_LL_{-1}+r_{GB}GB_{b-1})\}$
- $T_f=T_{ff}+T_{fb}$
- $S_w=-C_w+W-T_{iw}-T_{aw}-r_LL_{w-1}$
- $S_i=-C_i-T_{ii}-T_{ai}-T_{ci}+P_i+TD_b+r_{GB}GB_{i-1}$
- $NL_w=-C_{aw}+S_w$
- $NL_i=-C_{ai}+S_i$
- $NL_f=P_f-I$
- $NL_b=-W_b-T_{fb}-T_{cb}+P_b-TD_b+r_LL_{-1}+r_{GB}GB_{b-1}$
- $NL_g=-G-W_g+T_i+T_v+T_f+T_a+T_c-r_{GB}GB_{-1}$
- $L_w=\max\{\eta_1(W-T_{iw}-T_{aw}-r_LL_{w-1})\}$
- $\Delta L_w=L_w-L_{w-1}$
- $H_w=\eta_2C_w$
- $\Delta H_w=H_w-H_{w-1}$
- $\Delta M_w=NL_w+\Delta L_w-\Delta H_w$
- $M_w=M_{w-1}+\Delta M_w$
- $V_i^e=(1-\lambda_V)V_{i-1}^e+\lambda_VV_{i-1}\frac{V_{i-1}}{V_{i-1}-\Delta V_{i-1}}$
- $V_b^e=(1-\lambda_V)V_{b-1}^e+\lambda_VV_{b-1}\frac{V_{b-1}}{V_{b-1}-\Delta V_{b-1}}$
- $\Delta e=\frac{\eta_3(I-\beta_2K_{f-1})}{p_{e-1}}$
- $e=e_{-1}+\Delta e$
- $L_f=\eta_4(W_f+T_v+T_{ff}+T_{af})$
- $\Delta L_f=L_f-L_{f-1}$
- $\Delta L=\Delta L_w+\Delta L_f$
- $L=L_{-1}+\Delta L$
- $\Delta M_f=NL_f+\Delta L_f+p_{e-1}\Delta e$
- $M_f=M_{f-1}+\Delta M_f$

- $p_e=\frac{E_i^e+E_b^e}{e}$

- $H=H_{-1}+\Delta H$

- $GB_b=GB\frac{GB_b^e}{GB_i^e+GB_b^e}$
- $\Delta GB_i=GB_i-GB_{i-1}$
- $\Delta GB_b=GB_b-GB_{b-1}$

- $\Delta e_b=\Delta e-\Delta e_b$
- $e_i=e_{i-1}+\Delta e_i$
- $e_b=e_{b-1}+\Delta e_b$

- $E_i=E_{i-1}+\Delta E_i$

- $E=E_{-1}+\Delta E$

- $K_f=(1-\beta_2)K_{f-1}+I$
- $K_w=(1-\beta_4)K_{w-1}+C_{aw}$
- $K_i=(1-\beta_4)K_{i-1}+C_{ai}$

```julia
exogeneous = false
N=301

Cw, Ci, C, Caw, Cai, Ca, I, G = zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N)
W, Wf, Wb, Wg = zeros(N), zeros(N), zeros(N), zeros(N)
Tiw, Tii, Ti, Tv, Tff, Tfb, Tf, Taw, Tai, Taf, Ta, Tci, Tcb, Tc = zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N)
Pi, P, Pf, Pb, TDb, TDf, Sw, Si = zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N)
NLw, NLi, NLf, NLb, NLg = zeros(N), zeros(N), zeros(N), zeros(N), zeros(N)
ΔMw, ΔMi, ΔMf, ΔM = zeros(N), zeros(N), zeros(N), zeros(N)
ΔLw, ΔLf, ΔL = zeros(N), zeros(N), zeros(N)
Δei, Δeb, Δe = zeros(N), zeros(N), zeros(N)
ΔEi, ΔEb, ΔE, ΔNE = zeros(N), zeros(N), zeros(N), zeros(N)
D, ΔD = zeros(N), zeros(N)
ΔGBi, ΔGBb, ΔGB = zeros(N), zeros(N), zeros(N)
ΔHw, ΔHb, ΔH = zeros(N), zeros(N), zeros(N)
Kf, Kw, Ki, K = zeros(N), zeros(N), zeros(N), zeros(N)
Mw, Mi, Mf, M = zeros(N), zeros(N), zeros(N), zeros(N)
Lw, Lf, L = zeros(N), zeros(N), zeros(N)
ei, eb, e = zeros(N), zeros(N), zeros(N)
Ei, Eb, E, NE = zeros(N), zeros(N), zeros(N), zeros(N)
GBi, GBb, GB = zeros(N), zeros(N), zeros(N)
Hw, Hb, H = zeros(N), zeros(N), zeros(N)
NWw, NWi, NWf, NWb, NWg, NW = zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N)
Vi, Vb, Vie, Vbe, ΔVi, ΔVb = zeros(N), zeros(N), zeros(N), zeros(N), zeros(N), zeros(N)
Eie, GBie, Ebe, GBbe = zeros(N), zeros(N), zeros(N), zeros(N)
pe, Δpe = zeros(N), zeros(N)
rL, rGB, uT = 0.01, 0.02, 0.8
α1, α2, α3, α4, α5, α6 = 0.95, 0.05, 0.1, 0.8, 0.6, 0.01
β1, β2, β3, β4, β5 = 1, 0.05, 1, 0.05, 0.05
γ = 0.1
δ = 0.68
ϵ1, ϵ2, ϵ3, ϵ4, ϵ5, ϵ6, ϵ7, ϵ8, ϵ9 = 0.2, 0.1, 0.2, 0.4, 0.03, 0.03, 0.2, 0.01, 0.02
ζ1, ζ2 = 0.3, 0.1
η1, η2, η3, η4, η5 = 5, 0.05, 1, 1, 0.01
θ1, θ2 = 0.05, 0.3
λV, λw, λg = 0.5, 0.5, 0.55
λ10, λ20, λ30, λ40 = 0.3, 0.3, 0.1, 0.3
λ11, λ12, λ13, λ14 = 1, -0.3, -0.3, -0.4
λ21, λ22, λ23, λ24 = -0.3, 1, -0.4, -0.3
λ31, λ32, λ33, λ34 = -0.3, -0.4, 1, -0.3
λ41, λ42, λ43, λ44 = -0.4, -0.3, -0.3, 1
λ50, λ60 = 0.5, 0.5
λ51, λ52 = 0.5, -0.5
λ61, λ62 = -0.5, 0.5
G0, Wg0 = 1,0.9
G[1], Wg[1], W[1] = G0, Wg0, Wg0
pe[1] = 1
Kf[1], K[1] = 6, 6
Mw[1], Mi[1], Mf[1], M[1] = 1,1,1,3
ei[1], eb[1], e[1] = 2,2,4
Ei[1], Eb[1], E[1] = pe[1]*ei[1], pe[1]*eb[1], pe[1]*e[1]
D[1] = 1
GBi[1], GBb[1], GB[1] = 1,1,2
Hb[1], H[1] = η5*M[1],η5*M[1]
NWw[1] = Mw[1] - Lw[1] + Hw[1]
NWi[1] = Mi[1] + Ei[1] + D[1] + GBi[1]
NWf[1] = Mf[1] - Lf[1] - E[1] + Kf[1]
NWb[1] = -M[1] + L[1] + Eb[1] - D[1] + GBb[1] + Hb[1]
NWg[1] = -GB[1] - H[1]
NW[1] = K[1] + NE[1]
Vi[1], Vb[1], Vie[1], Vbe[1] = Mi[1]+Ei[1]+GBi[1], Eb[1]+GBb[1], Mi[1]+Ei[1]+GBi[1], Eb[1]+GBb[1]

function run(time_range::UnitRange)
    for t=time_range
        Wf[t] = (1-λw)*Wf[t-1] + λw*δ*(C[t-1]+Ca[t-1]+I[t-1]+G[t-1]-Tv[t-1]-Taf[t-1]-rL*Lf[t-1])
        Wb[t] = (1-λw)*Wb[t-1] + λw*δ*(-Tcb[t-1]+Pb[t-1]-TDb[t-1]+rL*L[t-1]+rGB*GBb[t-1])
        if exogeneous
            Wg[t] = Wg[t-1]*(1+γ)
        else
            Wg[t] = (1-λg)*(Ti[t-1]+Ta[t-1]+Tc[t-1]+Tv[t-1]+Tf[t-1]-rGB*(GB[t-1]-ΔGB[t-1]))*(1+γ)
        end
        W[t] = Wf[t] + Wb[t] + Wg[t]
        Tiw[t] = ϵ1*W[t]
        Tci[t] = max(0, ϵ7*Δpe[t-1]*(ei[t-1]-Δei[t-1])+ϵ9*(Ei[t-1]+D[t-1]))
        Tcb[t] = max(0, ϵ7*Δpe[t-1]*(eb[t-1]-Δeb[t-1])+ϵ9*Eb[t-1])
        Tc[t] = Tci[t] + Tcb[t]
        Taw[t] = ϵ5*Kw[t-1] + ϵ8*(Mw[t-1]-Lw[t-1])
        Tai[t] = ϵ5*Ki[t-1] + ϵ8*(Mi[t-1]+Ei[t-1]+GBi[t-1])
        Taf[t] = ϵ6*Kf[t-1]
        Ta[t] = Taw[t] + Tai[t] + Taf[t]
        if exogeneous
            G[t] = G[t-1]*(1+γ)
        else
            G[t] = λg*(Ti[t-1]+Ta[t-1]+Tc[t-1]+Tv[t-1]+Tf[t-1]-rGB*(GB[t-1]-ΔGB[t-1]))*(1+γ)
        end
        Cw[t] = α4*(α1*(W[t]-Tiw[t]-Taw[t]-rL*Lw[t-1]) + α2*(NWw[t-1]-Kw[t-1]))
        Ci[t] = α5*(α3*(Pi[t-1]+TDb[t-1]-Tii[t-1]-Tai[t-1]-Tci[t]+rGB*GBi[t-1])+α6*(Mi[t-1]+Ei[t-1]+D[t-1]+GBi[t-1]))
        C[t] = Cw[t] + Ci[t]
        Caw[t] = (1 - α4)*(α1*(W[t]-Tiw[t]-Taw[t]-rL*Lw[t-1]) + α2*(NWw[t-1]-Kw[t-1]))
        Cai[t] = (1 - α5)*(α3*(Pi[t-1]+TDb[t-1]-Tii[t-1]-Tai[t-1]-Tci[t]+rGB*GBi[t-1])+α6*(Mi[t-1]+Ei[t-1]+D[t-1]+GBi[t-1]))
        Ca[t] = Caw[t] + Cai[t]
        I[t] = max(0, min(β5*(Mf[t-1]-Lf[t-1])+((C[t]+Ca[t]+G[t])/(β1*Kf[t-1]) - uT)*Kf[t-1] + β2*Kf[t-1], β3*(Mf[t-1]-Lf[t-1])))
        Tv[t] = ϵ2*(C[t]+Ca[t]+G[t]+I[t])
        Tff[t] = max(0, ϵ3*(C[t]+Ca[t]+G[t]+I[t]-Wf[t]-Tv[t]-Taf[t]-rL*Lf[t-1]))
        P[t] = C[t]+Ca[t]+G[t]+I[t]-Wf[t]-Tv[t]-Tff[t]-Taf[t]-rL*Lf[t-1]
        TDf[t] = max(0, ζ1*(P[t]-I[t])+ζ2*(Mf[t-1]-Lf[t-1]))
        Pi[t] = TDf[t]*Ei[t-1]/(Ei[t-1]+Eb[t-1])
        Pb[t] = TDf[t]*Eb[t-1]/(Ei[t-1]+Eb[t-1])
        Pf[t] = P[t]-Pi[t]-Pb[t]
        TDb[t] = max(0, θ1*(-M[t-1]+L[t-1]+Eb[t-1]+GBb[t-1]+Hb[t-1]) + θ2*NLb[t-1])
        Tii[t] = ϵ4*(Pi[t]+TDb[t]+rGB*GBi[t-1])
        Ti[t] = Tiw[t] + Tii[t]
        Tfb[t] = max(0, ϵ3*(-Wb[t]-Tcb[t]+Pb[t]+rL*L[t-1]+rGB*GBb[t-1]))
        Tf[t] = Tff[t] + Tfb[t]
        Sw[t] = -Cw[t]+W[t]-Tiw[t]-Taw[t]-rL*Lw[t-1]
        Si[t] = -Ci[t]-Tii[t]-Tai[t]-Tci[t]+Pi[t]+TDb[t]+rGB*GBi[t-1]
        NLw[t] = -Caw[t]+Sw[t]
        NLi[t] = -Cai[t]+Si[t]
        NLf[t] = Pf[t] - I[t]
        NLb[t] = -Wb[t] - Tfb[t] - Tcb[t] + Pb[t] - TDb[t] + rL*L[t-1] + rGB*GBb[t-1]
        NLg[t] = -G[t] - Wg[t] + Ti[t] + Tv[t] + Tf[t] + Ta[t] + Tc[t] - rGB*GB[t-1]
        Lw[t] = max(0, η1*(W[t] - Tiw[t] - Taw[t] - rL*Lw[t-1]))
        ΔLw[t] = Lw[t] - Lw[t-1]
        Hw[t] = η2*Cw[t]
        ΔHw[t] = Hw[t] - Hw[t-1]
        ΔMw[t] = NLw[t] + ΔLw[t] - ΔHw[t]
        Mw[t] = Mw[t-1] + ΔMw[t]
        Vie[t] = (1-λV)*Vie[t-1] + λV*Vi[t-1]^2/(Vi[t-1]-ΔVi[t-1])
        Vbe[t] = (1-λV)*Vbe[t-1] + λV*Vb[t-1]^2/(Vb[t-1]-ΔVb[t-1])
        Δe[t] = η3*(I[t]-β2*Kf[t-1])/pe[t-1]
        e[t] = e[t-1] + Δe[t]
        Lf[t] = η4*(Wf[t]+Tv[t]+Tff[t]+Taf[t])
        ΔLf[t] = Lf[t] - Lf[t-1]
        ΔL[t] = ΔLw[t] + ΔLf[t]
        L[t] = L[t-1] + ΔL[t]
        ΔMf[t] = NLf[t] + ΔLf[t] + pe[t-1]*Δe[t]
        Mf[t] = Mf[t-1] + ΔMf[t]
        Mi[t] = (λ10 + λ12*TDf[t]/(Ei[t-1]+Eb[t-1]) + λ13*TDb[t]/D[t-1] + λ14*rGB)*Vie[t]
        Eie[t] = (λ20 + λ22*TDf[t]/(Ei[t-1]+Eb[t-1]) + λ23*TDb[t]/D[t-1] + λ24*rGB)*Vie[t]
        D[t] = (λ30 + λ32*TDf[t]/(Ei[t-1]+Eb[t-1]) + λ33*TDb[t]/D[t-1] + λ34*rGB)*Vie[t]
        GBie[t] = (λ40 + λ42*TDf[t]/(Ei[t-1]+Eb[t-1]) + λ43*TDb[t]/D[t-1] + λ44*rGB)*Vie[t]
        Ebe[t] = (λ50 + λ51*TDf[t]/(Ei[t-1]+Eb[t-1]) + λ52*rGB)*Vbe[t]
        GBbe[t] = (λ60 + λ61*TDf[t]/(Ei[t-1]+Eb[t-1]) + λ62*rGB)*Vbe[t]
        pe[t] = (Eie[t] + Ebe[t])/e[t]
        Δpe[t] = pe[t] - pe[t-1]
        ΔMi[t] = Mi[t] - Mi[t-1]
        ΔM[t] = ΔMw[t] + ΔMi[t] + ΔMf[t]
        M[t] = M[t-1] + ΔM[t]
        Hb[t] = η5*M[t]
        ΔHb[t] = Hb[t] - Hb[t-1]
        ΔH[t] = ΔHw[t] + ΔHb[t]
        H[t] = H[t-1] + ΔH[t]
        ΔGB[t] = -NLg[t] - ΔH[t]
        GB[t] = GB[t-1] + ΔGB[t]
        GBi[t] = GB[t]*GBie[t]/(GBie[t]+GBbe[t])
        GBb[t] = GB[t]*GBbe[t]/(GBie[t]+GBbe[t])
        ΔGBi[t] = GBi[t] - GBi[t-1]
        ΔGBb[t] = GBb[t] - GBb[t-1]
        Δei[t] = max(-ei[t-1], (NLi[t] - ΔMi[t] - ΔGBi[t])/pe[t-1])
        Δeb[t] = Δe[t] - Δei[t]
        ei[t] = ei[t-1] + Δei[t]
        eb[t] = eb[t-1] + Δeb[t]
        ΔEi[t] = Δpe[t]*ei[t] + pe[t-1]*Δei[t]
        ΔEb[t] = Δpe[t]*eb[t] + pe[t-1]*Δeb[t]
        Ei[t] = Ei[t-1] + ΔEi[t]
        Eb[t] = Eb[t-1] + ΔEb[t]
        ΔE[t] = pe[t-1]*Δe[t]
        E[t] = E[t-1] + ΔE[t]
        Vi[t] = Mi[t] + Ei[t] + D[t] + GBi[t]
        Vb[t] = Eb[t] + GBb[t]
        ΔD[t] = D[t] - D[t-1]
        Kf[t] = (1-β2)*Kf[t-1] + I[t]
        Kw[t] = (1-β4)*Kw[t-1] + Caw[t]
        Ki[t] = (1-β4)*Ki[t-1] + Cai[t]
        K[t] = Kf[t] + Kw[t] + Ki[t]
        NE[t] = Ei[t] + Eb[t] - E[t]
        ΔNE[t] = NE[t] - NE[t-1]
        NWw[t] = Kw[t] + Mw[t] - Lw[t] +Hw[t]
        NWi[t] = Ki[t] + Mi[t] + Ei[t] + D[t] + GBi[t]
        NWf[t] = Mf[t] - Lf[t] - E[t] + Kf[t]
        NWb[t] = -M[t] + L[t] + Eb[t] - D[t] + GBb[t] + Hb[t]
        NWg[t] = -GB[t] - H[t]
        NW[t] = K[t] + NE[t]
    end
end
run(2:N)
```

## 会計的一貫性の確認

- $NL_b+\Delta M-\Delta L-p_{e-1}\Delta e_b-\Delta GB_b-\Delta H_b=0$
- $\Delta NE=\Delta E_i-\Delta E+\Delta E_b$
- $M=M_w+M_i+M_f$
- $L=L_w+L_f$
- $GB=GB_i+GB_b$
- $H=H_w+H_b$
- $NW=NW_w+NW_i+NW_f+NW_b+NW_g$

の誤差をプロットする

```julia
using StatsPlots
plot(2:N, NLb[2:end]+ΔM[2:end]-ΔL[2:end]-pe[1:end-1].*Δeb[2:end]-ΔGBb[2:end]-ΔHb[2:end], label=nothing)
plot!(1:N, ΔNE+ΔE-ΔEi-ΔEb, label=nothing)
plot!(1:N, M-Mw-Mi-Mf, label=nothing)
plot!(1:N, L-Lw-Lf, label=nothing)
plot!(1:N, GB-GBi-GBb, label=nothing)
plot!(1:N, H-Hw-Hb, label=nothing)
plot!(1:N, NW-NWw-NWi-NWf-NWb-NWg, label=nothing)
```

いずれも計算機側の都合で誤差が出るが、のちにプロットする変数の値を見ればわかる通りオーダーが$10^{10}$以上違う。モデルの会計的一貫性は保たれていると言って良い
